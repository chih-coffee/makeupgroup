```python
content=list()
```

```python
from bs4 import BeautifulSoup
import requests
from time import sleep
from random import uniform
soup=list()
for i in range(5):
    url=f"https://www.ptt.cc/bbs/MakeUp/index{3850-i}.html"
    header={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}
    re=requests.get(url,headers=header)
    soup.append(BeautifulSoup(re.text,"html.parser"))
    sleep(uniform(0.4,1))
print(soup)
```

```python
for i in range(5):
    sections=soup[i].select('div[class="r-ent"]')
    keys=["[選擇]","[心得]","[妝容]"]
    for section in sections:
        num=section.find('div',{"class":"nrec"}).text
        if num and int(num)>=10:
            title=section.select('div[class="title"]')[0].text
            for key in keys:
                if key in title:
                    content.append([title.split(key)[1].replace("\n",""),int(num)])
print(content)
        

```

寫入檔案

```python
import csv
file=open("demo.csv","a+")
writor=csv.writer(file)
writor.writerows(content)
```
